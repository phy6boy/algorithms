<h2 style="color:purple"> Forward difference

The first principle formula for derivative is given by:
$$
f^\prime(x)\ =\ \lim_{h\rightarrow 0} \frac{f(x+h) - f(x)}{h}
$$

For computers to understand, The limit in the formula must be discarded. A reasonable approaximation is to take h as a small value near to 0 and write the formula as :
$$
f^\prime(x)\ =\ \frac{f(x+h) - f(x)}{h}, \quad 0 < h < \varepsilon
$$

This is a **forward difference formula**

**Question :** calculcate the derivative of $f(x) = e^x sin(x^2)$ in the interval $[-5, 5]$

```python
# loading necessary modules
import numpy as np                  # for numerical analysis
import matplotlib.pyplot as plt     # for plotting
```

```python
h = 1e-8                    # a small value (h)
N = 1000                    # number of samples in x axis
x = np.linspace(-5, 5, N)   # x axis (descetized)

def f(x):
    """
    The function on which the derivative is to be taken
    """
    return np.exp(x) * np.sin(x**2)

df = (f(x + h) - f(x))/h    # Taking derivative

# plotting values
plt.figure(figsize=(10, 8))
plt.plot(x, f(x), label="f(x)")
plt.plot(x, df  , label="f'(x)")
plt.legend()
plt.show()
```

```python
def df_true(x):
    """
    The true value of derivative
    """
    return np.exp(x) * (2*x*np.cos(x**2) + np.sin(x**2))

# plotting values
plt.figure(figsize=(10, 8))
plt.plot(x, df, label="f'(x) computed")
plt.plot(x, df_true(x), "o", label="f'(x) exact")
plt.legend()
plt.show()
```

<h2 style="color:purple;"> Central Difference

$$
f^\prime(x)\ =\ \frac{f(x+h) - f(x-h)}{2h}
$$

```python
df = (f(x + h) - f(x-h))/(2*h)    # Taking derivative

# plotting values
plt.figure(figsize=(10, 8))
plt.plot(x, f(x), label="f(x)")
plt.plot(x, df  , label="f'(x)")
plt.legend()
plt.show()
```
